- A. Conditional Statements (if, elif, else):

````python
num = 10
if num > 0:
    print(f"{num} is positive")
elif num == 0: # else if
    print(f"{num} is zero")
elif num == 5: # else if
    print(f"{num} is 5")
elif num == 0: # else if
    print(f"{num} is zero")
else:
    print(f"{num} is negative")
````

#### Control Flow Keywords:
- ```break:``` Terminates the loop prematurely.
- ```continue:``` Skips the current iteration of the loop.
- ```pass:``` Placeholder statement that does nothing (used for syntax completeness).

````python
# Calculating the sum of integers from 1 to 10
sum = 0
for i in range(1, 11):
    sum += i
print("Sum:", sum)  # Output: Sum: 55
````
