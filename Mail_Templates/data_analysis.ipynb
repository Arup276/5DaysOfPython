#### **Problem Statement:**
Given Sepal and Petal lengths and width predict the class of Iris

Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm
   5. classes: 
      - Iris Setosa
      - Iris Versicolour
      - Iris Virginica

#### Data cleaning and basic analysis

```python
# Load the data

df = pd.read_csv("Iris.csv")

df.head()
```

```python
# delete unwanted data

df = df.drop(columns= ['Id'])
df.head()
```

```python
# stats about the data

df.describe()
```

```python
# type of data

df.info()
```

#### Data visualization

```python
import matplotlib.pyplot as plt
```

```python
df['SepalLengthCm'].hist()
```

```python
df["SepalWidthCm"].hist()
```

```python
df['Species'].value_counts()
```

```python
# scatter plot

classes = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['orange', 'red', 'green']

for i in range(len(classes)):
    sps = df[df['Species'] == classes[i]]
    plt.scatter(sps['PetalLengthCm'], sps['PetalWidthCm'], c = colors[i], label = classes[i])

plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.legend()
```

```python
for i in range(len(classes)):
    sps = df[df['Species'] == classes[i]]
    plt.scatter(sps['SepalLengthCm'], sps['SepalWidthCm'], c = colors[i], label = classes[i])

plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.legend()
```
